```python
import pandas as pd
import numpy as np
```

```python
credit = pd.read_csv("tmdb_5000_credits.csv")
movie = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movie.head(10)
```

```python
movie.shape
```

```python
credit.head(10)
```

```python
movie = movie.merge(credit ,on='title')
```

```python
movie.head(10)
```

```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movie.dropna(inplace=True)
```

```python
movie['genres'] = movie['genres'].apply(convert)
movie.head()
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
movie.head()
```

```python
movie['cast'] = movie['cast'].apply(convert)
movie.head()
```

```python
movie['cast'] = movie['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    EL = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            EL.append(i['name'])
    return EL
```

```python
movie['crew'] = movie['crew'].apply(fetch_director)
```

```python
movie.sample(10)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movie['cast'] = movie['cast'].apply(collapse)
movie['crew'] = movie['crew'].apply(collapse)
movie['genres'] = movie['genres'].apply(collapse)
movie['keywords'] = movie['keywords'].apply(collapse)
```

```python
movie.head(10)
```

```python
movie.columns
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie['tags'] = movie['overview'] + movie['genres'] + movie['keywords'] + movie['cast'] + movie['crew']
```

```python
new = movie.drop(columns=['overview','genres','keywords','cast','crew','budget','genres','homepage','id','original_language','original_title','popularity','production_companies','production_countries','release_date','revenue','runtime','spoken_languages','status','tagline','vote_average','vote_count'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new.to_csv('data.csv', index=False)
```
